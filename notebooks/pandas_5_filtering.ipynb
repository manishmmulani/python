```python
import pandas as pd
```

```python
imdb = pd.read_csv('http://bit.ly/imdbratings')
```

```python
imdb.head()
```

## Filtering on column names or row index not on content

```python
imdb.filter(items=['title', 'star_rating']).head()
```

```python
imdb.filter(like='rating').head()
```

```python
imdb.filter(regex='e$', axis=1).head()
```

```python
imdb.filter(items=[0,1,2], axis=0).head()
```

## Filtering on content

```python
imdb["star_rating"].head()
```

```python
(imdb["star_rating"] > 9.0).head()
```

```python
imdb[imdb["star_rating"] > 8.5].head()
```

```python
imdb[imdb["star_rating"] > 9.0]["title"]
```

```python
# The above to select only the titles is not the best practice. Instead use loc as below
# first part is row filter, second part is column filter
```

```python
imdb.loc[imdb["star_rating"] > 9.0, "title"]
```

## Filtering based on String methods on dataframes

```python
imdb.genre.str.lower().head()
```

```python
imdb[imdb.title.str.contains("The")].head()
```

## Filtering by multiple criteria - use parenthesis and spearate by & or |

```python
imdb[(imdb["star_rating"] > 9.0) & (imdb["genre"] == 'Crime')]
```

```python
# isin for multiple ORs
imdb[imdb.genre.isin(['Crime', 'Drama', 'Action'])].head()
```

```python
# the above is same as
imdb[(imdb["genre"] == 'Crime') | (imdb["genre"] == 'Drama') | (imdb["genre"] == 'Action')].head()
```
